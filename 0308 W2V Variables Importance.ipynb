```python
df_w2v = pd.read_csv('kyrbs2019_W2V.csv')
df_w2v.drop(columns = 'Unnamed: 0', axis = 1 ,inplace = True)
```

```python
#create Vocab dict
cols = []
for j in df_w2v.columns:
    target = df_w2v[j].unique()
    for i in target:
        cols.append(i)
```

```python
# c1  Yes , c2  No 
def vec_means(cols, model):
    c1 = model.wv['BMI_1']
    c2 = model.wv['BMI_0']
    c1_cluster = []
    c2_cluster = []
    for col in cols:
        temp = []
        if np.linalg.norm(c1 - model.wv[col]) > np.linalg.norm(c2 - model.wv[col]): #    
            temp.append(col)
            temp.append(np.linalg.norm(c2 - model.wv[col]))
            c2_cluster.append(temp)
        else:
            temp.append(col)
            temp.append(np.linalg.norm(c1 - model.wv[col]))
            c1_cluster.append(temp)
            
    return c1_cluster, c2_cluster
```

```python
c1_c , c2_c = vec_means(cols, model_sg)
```

```python
c1_c = pd.DataFrame(c1_c)
c2_c = pd.DataFrame(c2_c)
```

```python
def variables_ed(variable):
    '''           Yes   , 
       No              '''
    biman = []
    no_biman = []
    ED = []
    for value in list(set(df_w2v[variable].values)):
        if value in c1_atts:
            biman.append(value)
        elif value not in c1_atts:
            no_biman.append(value)
    for i in biman:
        for j in no_biman:
            ED.append(distance.euclidean(emb_vec[i], emb_vec[j]))
            
    return min(ED)
#    return((min(list(set(ED)))))
```

```python
def variables_ed_max(variable):
    '''           Yes   , 
       No              '''
    biman = []
    no_biman = []
    ED = []
    for value in list(set(df_w2v[variable].values)):
        if value in c1_atts:
            biman.append(value)
        elif value not in c1_atts:
            no_biman.append(value)
    for i in biman:
        for j in no_biman:
            ED.append(distance.euclidean(emb_vec[i], emb_vec[j]))
            
    return((max(list(set(ED)))))
```
