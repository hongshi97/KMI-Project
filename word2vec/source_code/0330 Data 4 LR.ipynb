```python
# 데이터셋 불러오기
df_og = pd.read_csv('kyrbs2019_boy.csv')
df_og.drop('Unnamed: 0', axis = 1, inplace=True)
df_og
```

```python
df_min = pd.read_csv('df_min.csv', index_col = 0)
df_max = pd.read_csv('df_max.csv', index_col = 0)
```

```python
df = df_og.copy()
```

```python
# 최소 ED로 추출한 변수           변수 순서 바꾸지 말기!!! WC_MN_M10부터가 변수 중요도 1순위였음
att_min = df_min.Variables.tolist()
df_min_ED = df.loc[:,att_min]
```

```python
df_min_ED
```

```python
# 최대 ED로 추출한 변수  ( 변수 순서 바꾸지 말기!!! E_BORN_M 부터가 변수 중요도 1순위였음)
df_max_ED = df[df_max.Variables.tolist()]
```

```python
df_max_ED
```

## df_min_ED, df_max_ED 데이터 다중공선성 확인하기

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
# 다중공선성 확인
def check_vif(df):
    global vif
    vif = pd.DataFrame()
    vif["VIF Factor"] = [variance_inflation_factor(
        df.values, i) for i in range(df.shape[1])]
    vif["features"] = df.columns
```

```python
# df_min 에서 VIF 확인 후 변수 제거
for i in range(1000):
    check_vif(df_min_ED)
    if len(vif[vif['VIF Factor'] >=10]) >=1:
        remove_col = vif[vif['VIF Factor'] >=10].sort_values('VIF Factor', ascending=False).iloc[0,1]
        df_min_ED.drop(remove_col, axis=1, inplace=True)
        print('제거된 변수는', remove_col)
    else:
        display(vif)
        break
```

```python
vif.features.tolist()
```

```python
len(vif.features.tolist())
```

```python
# df_min_ED에서 VIF 확인 후 살아남은 변수들 + 비만 여부 변수 ( Target Attribute )만 다시 df에서 뽑아서 df_min_ED에 할당
df_min_ED = df[vif.features.tolist() + ['BMI']]
```

```python
df_min_ED
```

```python
# df_max 에서 VIF 확인 후 변수 제거
for i in range(1000):
    check_vif(df_max_ED)
    if len(vif[vif['VIF Factor'] >=10]) >=1:
        remove_col = vif[vif['VIF Factor'] >=10].sort_values('VIF Factor', ascending=False).iloc[0,1]
        df_max_ED.drop(remove_col, axis=1, inplace=True)
        print('제거된 변수는', remove_col)
    else:
        display(vif)
        break
```

```python
vif.features.tolist()
```

```python
len(vif.features.tolist())
```

```python
# df_max_ED에서 VIF 확인 후 살아남은 변수들 + 비만 여부 변수 ( Target Attribute )만 다시 df에서 뽑아서 df_max_ED에 할당
df_max_ED = df[vif.features.tolist() +['BMI']]
```

```python
# 데이터 저장
df_min_ED.to_csv('df_min_ED.csv')
df_max_ED.to_csv('df_max_ED.csv')
```

## LR용 Train/Test Set 만들기

```python
from sklearn.model_selection import train_test_split

X = df_min_ED.iloc[:,:-1]
y = df_min_ED[['BMI']]

min_X_train, min_X_test, min_y_train, min_y_test = train_test_split(X,y, test_size = 0.3, shuffle = True, stratify = y, random_state= 1002)
```
