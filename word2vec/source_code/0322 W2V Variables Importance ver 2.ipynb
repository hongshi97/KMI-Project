```python
df_w2v = pd.read_csv('kyrbs2019_W2V.csv')
df_w2v.drop(columns = 'Unnamed: 0', axis = 1 ,inplace = True)
```

```python
cols = []
for temp in df_w2v.columns:
    target = df_w2v[temp].unique()
    for a in target:
        cols.append(a)
```

```python
len(cols)
```

```python
cos = []
for col in cols:
    cos.append(model_sg.wv.similarity(col,'BMI_0'))
```

```python
cos
```

```python
cos2 = []
for col in cols:
    cos2.append(model_sg.wv.similarity(col, 'BMI_1'))
```

```python
cos2
```

```python
temp_df = pd.concat([pd.Series(cols),pd.Series(cos)], axis=1)
print(temp_df)
```

```python
temp_df = temp_df.sort_values(by = 1, ascending = False)
temp_df.reset_index(drop = True, inplace = True)
```

```python
temp_df.columns = ['Factors','BMI_0']
```

```python
temp_df['Rank_BMI_0'] = pd.Series(list(range(len(temp_df))))
```

```python
def cosine_mean(att):
    list_a = []
    for value in list(b.index):
        if att in value:
            list_a.append(b.loc[value,'cosine difference'])
    return np.mean(np.array(list_a))
```

```python
list_b = []
for att in atts:
    list_b.append(cosine_mean(att))
len(list_b)
```
