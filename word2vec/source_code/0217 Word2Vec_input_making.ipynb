```python
import pandas as pd
```

```python
# 데이터셋 불러오기
df_og = pd.read_csv('kyrbs2019_boy.csv')
df_og.drop('Unnamed: 0', axis = 1, inplace=True)
df_og
```

```python
# 분석용 데이터셋 생성(원본 데이터 복제)
df = df_og.iloc[:,:]
df
```

```python
df.BMI.value_counts()
```

```python
for i in list(df.columns):
    print(df[i].unique)
```

```python
# M_SAD_1 처럼 변수값 바꾸기
for i in list(df.columns):
    for j in df[i].unique():
        df[[i]] = df[[i]].replace({j:i+"_"+str(int(j))})
```
