```
import sentencepiece as spm
VOC_SIZE = 141
NUM_PLACEHOLDERS = 10
spm.SentencePieceTrainer.Train(input="/content/drive/My Drive/bert_suicide2/nlp data.csv", model_prefix='khuspm',max_sentencepiece_length=231, model_type ='word',vocab_size=VOC_SIZE-NUM_PLACEHOLDERS)
```

```
def read_sentencepiece_vocab(filepath):
  voc = []
  with open(filepath, encoding='utf-8') as fi:
    for line in fi:
      voc.append(line.split("\t")[0])
  # skip the first <unk> token
  voc = voc[1:]
  return voc

khu_vocab = read_sentencepiece_vocab("{}.vocab".format("khuspm"))
print("Learnt vocab size: {}".format(len(khu_vocab)))
```

```
khu_vocab
```

```
def parse_sentencepiece_token(token):
    if token.startswith("▁"):
        return token[1:]
    else:
        return "##" + token

bert_vocab = list(map(parse_sentencepiece_token, khu_vocab))
ctrl_symbols = ["[PAD]","[UNK]","[CLS]","[SEP]","[MASK]"]
bert_vocab = ctrl_symbols + bert_vocab
print(len(bert_vocab))
```

```
bert_vocab += ["[UNUSED_{}]".format(i) for i in range(VOC_SIZE - len(bert_vocab))]
print(len(bert_vocab))
```

```
VOC_FNAME = "vocab.txt" #@param {type:"string"}

with open(VOC_FNAME, "w") as fo:
  for token in bert_vocab:
    fo.write(token+"\n")
```
